```python
#Task 1
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
import matplotlib.pyplot as plt

df=pd.read_csv("D:/ai workforce/aug 23 practice/svm_linear.csv")
df.head()
```

```python
#i have 3 columns.using iloc separating the features and target
x=df.iloc[:,:-1]     # [all rows,2 columns]
y=df.iloc[:,-1]      #[all rows,last column]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=SVC(kernel='linear')     #question the kernel is linear
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
ac=accuracy_score(y_test,y_pred)
ac            #accu close to 1 means satisfying
```

```python
#plot the poits before decison boundary
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=y,cmap='bwr',edgecolors='black')
```

```python
#decision boundary   formula =w1x1+w2x2+b=0
w=model.coef_[0]
b=model.intercept_[0]
s=model.support_vectors_
s
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values, y.values, clf=model, legend=2)
```

```python
#Task 4
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
df=pd.read_csv("D:/ai workforce/aug 23 practice/dt_class.csv")
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Task 5
df=pd.read_csv("D:/ai workforce/aug 23 practice/dt_depth.csv")
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
for depth in [2,5,None]:  #here none means full depth
    model=DecisionTreeClassifier(max_depth=depth,random_state=42)    #if depth not defined,,it may completely train and test result in overfitting
    model.fit(x_train,y_train) 
    train_ac=model.score(x_train,y_train)
    test_ac=model.score(x_test,y_test)
    print(f"Depth={depth}      train ac= {train_ac}    test_ac={test_ac}")
```

```python
#Task 6
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
df=pd.read_csv("D:/ai workforce/aug 23 practice/dt_reg.csv")
df.head()
```

```python
X = df[['x']].values
y=df['y'].values
```

```python
from sklearn.metrics import mean_squared_error
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)
```

```python
for depth in [2, 5, None]:
    model = DecisionTreeRegressor(max_depth=depth, random_state=42)
    model.fit(x_train, y_train)
    train_rmse = np.sqrt(mean_squared_error(y_train, model.predict(x_train)))
    test_rmse  = np.sqrt(mean_squared_error(y_test, model.predict(x_test)))
    print(f"Max Depth={depth} → Train RMSE: {train_rmse:.3f}, Test RMSE: {test_rmse:.3f}")
```
